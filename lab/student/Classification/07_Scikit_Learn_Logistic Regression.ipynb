# 使用 Scikit-Learn 進行邏輯回歸



## 目標
在這個實驗中，你將會：
- 使用 scikit-learn 訓練一個邏輯回歸模型。


## 資料集
我們先沿用和前面相同的資料集。

```python
import numpy as np

X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])
```

## 擬合模型

下方程式碼會從 scikit-learn 匯入 [邏輯回歸模型](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression)。你可以呼叫 `fit` 來使用訓練資料完成模型擬合。

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X, y)
```

## 進行預測

你可以呼叫 `predict` 來查看模型在資料上的預測結果。

```python
y_pred = lr_model.predict(X)

print("Prediction on training set:", y_pred)
```

## 計算準確率

你可以呼叫 `score` 來計算模型的準確率（accuracy）。

```python
print("Accuracy on training set:", lr_model.score(X, y))
```
